```python
import pandas as pd

# Load the uploaded Excel file
file_path = 'komentar seller.xlsx'
data = pd.read_excel(file_path)

# Show the first few rows of the dataset
data.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# Extract the comment column (last column)
comments = data.iloc[:, -1].dropna()

# Define a function to generate n-grams (n=2 for this example)
def generate_ngrams(comments, n=2):
    vectorizer = CountVectorizer(ngram_range=(n, n), token_pattern=r'\b\w+\b', min_df=1)
    X = vectorizer.fit_transform(comments)
    ngrams = vectorizer.get_feature_names_out()
    return ngrams

# Generate bigrams (n=2) for the comments
bigrams = generate_ngrams(comments, n=2)

# Display the bigrams
bigrams[:20]  # Display first 20 ngrams for brevity
```

```python
import pandas as pd
from nltk import ngrams

# Load the dataset
file_path = 'komentar seller.xlsx'
data = pd.read_excel(file_path)

# Ambil kolom komentar yang tampaknya adalah kolom ke-4
comments = data.iloc[:, 3].tolist()  # Mengambil kolom ke-4 yang berisi komentar

# Fungsi untuk membuat N-grams
def generate_ngrams(text, n=2):
    words = text.split()
    return list(ngrams(words, n))

# Iterasi melalui komentar dan cetak N-grams dengan format tuple
for i, comment in enumerate(comments):
    print(f"Komentar ke-{i+1}")
    print("_" * 30)  # Garis bawah pemisah
    ngrams_output = generate_ngrams(comment, n=2)  # Menghasilkan bigram (N-grams dengan n=2)
    for gram in ngrams_output:
        print(gram)  # Menampilkan N-grams dalam format tuple
    print("\n")  # Baris baru untuk memisahkan setiap komentar
```
